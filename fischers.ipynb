```python
classes=[c0,c1,c2,c3,c4,c5,c6,c7,c8,c9]
notclasses=[notc0,notc1,notc2,notc3,notc4,notc5,notc6,notc7,notc8,notc9]
```

```python
mean0=np.sum(notc0,axis=0)/len(notc0)
mean1=np.sum(notc1,axis=0)/len(notc1)
mean2=np.sum(notc2,axis=0)/len(notc2)
mean3=np.sum(notc3,axis=0)/len(notc3)
mean4=np.sum(notc4,axis=0)/len(notc4)
mean5=np.sum(notc5,axis=0)/len(notc5)
mean6=np.sum(notc6,axis=0)/len(notc6)
mean7=np.sum(notc7,axis=0)/len(notc7)
mean8=np.sum(notc8,axis=0)/len(notc8)
mean9=np.sum(notc9,axis=0)/len(notc9)

meanc0=np.sum(c0,axis=0)/len(c0)
meanc1=np.sum(c1,axis=0)/len(c1)
meanc2=np.sum(c2,axis=0)/len(c2)
meanc3=np.sum(c3,axis=0)/len(c3)
meanc4=np.sum(c4,axis=0)/len(c4)
meanc5=np.sum(c5,axis=0)/len(c5)
meanc6=np.sum(c6,axis=0)/len(c6)
meanc7=np.sum(c7,axis=0)/len(c7)
meanc8=np.sum(c8,axis=0)/len(c8)
meanc9=np.sum(c9,axis=0)/len(c9)
```

```python
means1=[mean0,mean1,mean2,mean3,mean4,mean5,mean6,mean7,mean8,mean9]#deosnt belong
means2=[meanc0,meanc1,meanc2,meanc3,meanc4,meanc5,meanc6,meanc7,meanc8,meanc9]#belongs
```

```python
SWj1=[]
for i in range(10):
  sum=np.zeros((784,784))
  for j in range(240):
    x1= (classes[i][j]-means2[i]).reshape(784,1)
    xtrans1=x1.reshape(1,784)
    xxt1=x1.dot(xtrans1)
    sum=sum+xxt1

  SWj1.append(sum)
```

```python
SWj2=[]
for i in range(10):
  sum=np.zeros((784,784))
  for j in range(2160):
    x1= (notclasses[i][j]-means1[i]).reshape(784,1)
    xtrans1=x1.reshape(1,784)
    xxt1=x1.dot(xtrans1)
    sum=sum+xxt1

  SWj2.append(sum)
```

```python
SW=[]
for i in range(10):
 sw=SWj1[i]+SWj2[i]
 SW.append(sw)
```

```python
SWINV=[]
for i in range(10):
 inverse=np.linalg.pinv(SW[i])
 SWINV.append(inverse)
```

```python
weights=[]
for i in range(10):
  w=(means1[i]-means2[i]).dot(SWINV[i])
  weights.append(w)

```

```python
result=np.ones([200])
for i in range(200):
  image=Test[i]
  tests=np.zeros([10])
  for j in range(10):
   currw=weights[j]
   tests[j]=image.dot(currw)
  min=np.argmin(tests)
  result[i]= min

```
